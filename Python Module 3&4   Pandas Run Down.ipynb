# Treating missing values in python

```python
# creating 6 x 6 dataframe with randon numbers
np.random.seed(25)
df_obj2 = DataFrame(np.random.rand(36).reshape((6,6)), index=['row 1', 'row 2','row 3', 'row 4','row 5', 'row 6'], columns=['col 1', 'col 2','col 3', 'col 4', 'col 5', 'col 6'])

df_obj_missing
```

```python
#using fillna
df_obj_missing.fillna(0.5)
```

```python
df_obj_missing.fillna(method='ffill')
```

```python
df_obj_missing.fillna(df_obj_missing.mean())
```

```python
#fill with predefined colum dict
col3mean = np.mean(df_obj_missing.loc[:,'col 3'])
df_obj_missing.fillna({'col 2':col3mean,'col 5':.5})
```

```python
#fillna method = forward fill
df_obj2.fillna(method='ffill')
```

```python
# method fillna mean ,, there is no such method
#df_obj2.fillna(method='mean')
```

```python
#we can do by calling pandas mean to fillna
df_obj2.fillna(df_obj2.mean())
```

```python
#rounding and formating
df_obj2 = df_obj2.round(3)*100
df_obj2
```

```python
#removing records with missing values
df_obj_missing
```

```python
df_obj_missing.dropna(axis=1)
```

```python
#removing records with na colums wise(not useful normally), column axis=1, row axis=0
df_obj2.dropna(axis=0)
```

```python
#remove row with all na
print(df_obj_missing.dropna(how="all"))
```

# Grouping and Aggregation 

```python
#grouping using mtcars data set

file_path ="mtcars.csv"
cars = pd.read_csv(file_path)
cars.head()
```

```python
#assign column names  ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb'] using columns att
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb'] 
cars.head()
```

```python
#grouping
cars_groups = cars.groupby(cars['am'])
print(cars_groups)
cars_groups.mean()
```
