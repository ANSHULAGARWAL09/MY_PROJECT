<h3>Split Train Test</h3>

```python
# Split sample into Train and Test
Train,Test = train_test_split(binary, test_size = 0.3, random_state = 17)
# Print a few rows
Train.head()
```

```python
# Split Target and Feature Set
# Keep Target and Independent Variable into different array
Train_IndepentVars = Train.values[:,3:5]
Train_TargetVar = Train.values[:,5]
Train_IndepentVars
```

<h3>Random Forest Model</h3>

```python
rf_model =  RandomForestClassifier(random_state = 42)
rf_model.fit(Train_IndepentVars,Train_TargetVar.astype(int))
```

```python
# Scoring based on the train RF Model
predictions = rf_model.predict(Train_IndepentVars)
```

```python
# Confusion Matrix
print(confusion_matrix(Train_TargetVar.astype(int), predictions))
```

```python
print(accuracy_score(Train_TargetVar.astype(int), predictions))
```
