Since we are given life expectancy level data by country and year, we could ask questions about how much the average life expectancy changes over each year.

Before continuing, however, it's important to make sure that the following assumptions about the data are true:

>'Life expectancy' is the first column (index 0) of the DataFrame.
> The other columns contain either null or numeric values.
> The numeric values are all greater than or equal to 0.
> There is only one instance of each country.

We can write a function that can apply over the entire DataFrame to verify some of these assumptions. 

```python
def check_null_or_valid(row_data):
    """Function that takes a row of data,
    drops all missing values,
    and checks if all remaining values are greater than or equal to 0
    """
    
    no_na = row_data.dropna()[1:-1]
    numeric = pd.to_numeric(no_na)
    ge0 = numeric >= 0
    return ge0

# Check whether the first column is 'Life expectancy'
assert dataset.columns[0] == 'Life expectancy'

# Check whether the values in the row are valid
assert dataset.iloc[:, 1:].apply(check_null_or_valid, axis=1).all().all()

# Check that there is only one instance of each country
assert dataset['Life expectancy'].value_counts()[0] == 1
```

Here, the error is expected as according to us 'Life expectancy' column should be there but data is not having this column as the first one. This is the reason of error. 

The next step is to reshape it into a tidy data format.

The gapminder DataFrame has a separate column for each year. What we want instead is a single column that contains the year, and a single column that represents the average life expectancy for each year and country. By having year in its own column, we can use it as a predictor variable in a later analysis.

We can convert the DataFrame into the desired tidy format by melting it.

```python
# Melt gapminder: gapminder_melt
dataset_melt = pd.melt(frame = dataset, id_vars='Life expectancy')

# Rename the columns
dataset_melt.columns = ['country', 'year', 'life_expectancy']

# Print the head of gapminder_melt
print(dataset_melt.head())
```

Having the data in this tidy format will make subsequent analysis far easier.

Now that the data is in the proper shape, we need to ensure that the columns are of the proper data type. That is, we need to ensure that country is of type object, year is of type int64, and life_expectancy is of type float64.

```python
dataset_melt
```

```python
# Convert the year column to numeric
dataset_melt.year = pd.to_numeric(dataset_melt.year)

# Test if country is of type object
assert dataset_melt.country.dtypes == np.object

# Test if year is of type int64
assert dataset_melt.year.dtypes == np.int64

# Test if life_expectancy is of type float64
assert dataset_melt.life_expectancy.dtypes == np.float64
```
