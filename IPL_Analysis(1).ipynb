```python
fielding = matches[matches['TOSS_DECISION'] == 'field']
```

```python
fielding_16 = fielding[fielding['SEASON'] == 2016]
```

```python
fielding_17 = fielding[fielding['SEASON'] == 2017]
```

```python
grp_field16 = fielding_16.groupby('TOSS_WINNER').count()
grp_field16['COUNT'] = grp_field16.iloc[:,1]
```

```python
grp_field17 = fielding_17.groupby('TOSS_WINNER').count()
grp_field17['COUNT'] = grp_field17.iloc[:,1]
```

```python
grp_field16 = grp_field16.sort_values(by='COUNT',ascending=False)
grp_field16 = pd.DataFrame(grp_field16.iloc[:2,-1])
grp_field16
```

```python
grp_field16.reset_index(inplace=True)
```

```python
grp_field16['YEAR'] = 2016
```

```python
grp_field16.rename(columns={'TOSS_WINNER':'TEAM'},inplace=True)
```

```python
grp_field16
```

```python
grp_field17 = grp_field17.sort_values(by='COUNT',ascending=False)
grp_field17 = pd.DataFrame(grp_field17.iloc[:2,-1])
grp_field17
```

```python
grp_field17.reset_index(inplace=True)
```
